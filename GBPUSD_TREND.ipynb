```python
import yfinance as yf
import matplotlib.pyplot as plt

# Define the ticker symbol
ticker = "GBPUSD=X"

# Define the August periods you want to analyze
years = [2024, 2023, 2022, 2021, 2020, 2019, 2018, 2017]

# Initialize a dictionary to store the data
data_dict = {}

# Fetch the historical data for each year
for year in years:
    start_date = f"{year}-08-01"
    end_date = f"{year}-08-31"
    data = yf.download(ticker, start=start_date, end=end_date)
    data_dict[year] = data

# Plotting the data for comparison
plt.figure(figsize=(14, 8))

# Iterate through each year's data to plot
for year in years:
    plt.plot(data_dict[year].index, data_dict[year]['Close'], label=f'August {year}', marker='o', linestyle='-')

# Add title and labels
plt.title('GBP/USD Closing Prices Comparison: August 2017 - August 2024')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend(title="Year")
plt.tight_layout()

# Show the plot
plt.show()
```
