```python
import yfinance as yf
import pandas as pd

# Define the ticker symbol and the date range for August 2022
ticker = "GBPUSD=X"
start_date = "2022-08-01"
end_date = "2022-08-31"

# Fetch the historical data for GBP/USD for August 2022
data = yf.download(ticker, start=start_date, end=end_date)

# Save the data to a CSV file
#data.to_csv("gbpusd_august_2023.csv")
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['Close'], marker='o', linestyle='-', color='b')
plt.title('GBP/USD Closing Prices - August 2022')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```
